# Explanation of Python Imports

## 1. `import pandas as pd`
- **Pandas** is a powerful data analysis library. It allows you to create and manipulate data tables (DataFrames) effortlessly, making it great for data analysis and cleaning.

## 2. `import numpy as np`
- **NumPy** is essential for numerical and mathematical operations. It helps you work with arrays and perform complex calculations efficiently.

## 3. `import matplotlib.pyplot as plt`
- **Matplotlib** is a popular library for creating plots and charts. The `pyplot` module (`plt`) is used to make simple visualizations, like line plots and bar charts, to better understand your data.

## 4. `import seaborn as sns`
- **Seaborn** extends Matplotlib’s functionality, making it easier to create visually appealing and informative statistical charts and plots.

## 5. `from sklearn.tree import DecisionTreeRegressor`
- This imports a **Decision Tree Regressor** model from `scikit-learn`. It’s used for predicting continuous values by splitting data into branches based on features.

## 6. `from sklearn.linear_model import LinearRegression, Lasso, Ridge`
- These are different types of regression models:
  - **`LinearRegression`**: A standard linear model for predicting continuous outputs.
  - **`Lasso`**: A model with L1 regularization that helps reduce overfitting by zeroing out less important features.
  - **`Ridge`**: A model with L2 regularization that limits large coefficients to make the model more general.

## 7. `from sklearn.preprocessing import StandardScaler`
- `StandardScaler` standardizes data by removing the mean and scaling to unit variance. This ensures that all features contribute equally to the model's performance.

## 8. `from sklearn.ensemble import RandomForestRegressor`
- **Random Forest Regressor** is an ensemble model that uses multiple decision trees to improve prediction accuracy and reduce overfitting by averaging the outputs of several trees.

## 9. `from sklearn.metrics import mean_squared_error, mean_absolute_error`
- These metrics are used to evaluate model performance:
  - **`mean_squared_error (MSE)`**: Shows the average squared difference between actual and predicted values, with larger errors penalized more.
  - **`mean_absolute_error (MAE)`**: Shows the average absolute difference between actual and predicted values, making it easier to interpret.

## 10. `from xgboost import XGBRegressor`
- **XGBoost** is a highly efficient and powerful boosting algorithm used for regression tasks. It's known for its speed and accuracy and is often used in machine learning competitions.

## 11. `from sklearn.model_selection import train_test_split, cross_val_score`
- Utilities for splitting and evaluating your data:
  - **`train_test_split`**: Splits your data into training and testing sets to evaluate model performance.
  - **`cross_val_score`**: Performs cross-validation to assess model performance on multiple data subsets, ensuring reliability.

## 12. `import math`
- The **math** library provides basic mathematical functions, like square roots and logarithms, useful for calculations during data analysis.

## 13. `from sklearn.preprocessing import LabelEncoder`
- `LabelEncoder` transforms categorical text data into numeric labels so that machine learning models can understand and use them.

These imports are essential for building, visualizing, and evaluating machine learning models on Kaggle, helping you go from data preparation to model validation with ease.

# **PIECHART for cat Data**

```python
n_cols = len(cat.columns)
fig, axes = plt.subplots(nrows=2 , ncols=2, figsize=(10, 8))  
axes = axes.flatten()

for i, col in enumerate(cat.columns):
    x = cat[col].value_counts()
    axes[i].pie(x, labels=x.index, autopct='%1.1f%%' , wedgeprops={"linewidth":2 , "edgecolor": "white"})
    axes[i].set_title(f"Distribution of {col}")

plt.tight_layout()
plt.show()
```

# **Encodeing**

SEX

```python
gender = {'male' : 0 , 'female' : 1}
data['sex'] = data['sex'].apply(lambda x : gender[x] )
data.head()
```

SMOKERS(YES,NO)(1,0)

```python
smokers = {'no' : 0 , 'yes':1 }
data['smoker'] = data['smoker'].apply(lambda x : smokers[x])
data.head()
```

```python
data['region'].unique()
```

Region column to be 0,1,2,3

```python
regions = {'southwest' : 0, 'southeast' : 1, 'northwest' : 2 , 'northeast' : 3}
data['region'] = data['region'].apply(lambda x : regions[x])
data.head()
```
